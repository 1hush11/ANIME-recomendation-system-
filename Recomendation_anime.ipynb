### Коллаборативная фильтрация с ипользованием корреляции

```python
# создание случайной выборки из 1 млн строк
new_df_sample = df.sample(n=1000000, random_state=1234)
```

```python
# сортировка выборки, где 'user_id' встречается в датасете больше 100 раз
rec_df = new_df_sample[new_df_sample['user_id'].map(new_df_sample['user_id'].value_counts()) > 100]  
```

```python
rec_df
```

```python
# создание размеренной таблицы
new_pivot_table = rec_df.pivot_table(index=["user_id"],columns=["name"],values="user rating")
new_pivot_table.fillna(0,inplace=True)
```

```python
# корелляционная функция для поиска похожего аниме
def same_anime(anime_name, threshold=0.8, top_n=10):
    # проверка наличия указанного аниме в таблице
    if anime_name not in new_pivot_table.columns:
        print(f"'{anime_name}' не найдено в базе данных.")
        return pd.DataFrame()
    
    # данные оценок пользователей для указанного аниме
    users_vote_anime = new_pivot_table[anime_name]
    
    # Вычисляем корреляцию с другими аниме, игнорируя отсутствующие значения
    similar_with = new_pivot_table.corrwith(users_vote_anime, method='pearson')
    similar_with = pd.DataFrame(similar_with, columns=['correlation']).dropna()
    
    # сортировка и фильтрация по порогу корреляции
    df_similar = similar_with.sort_values('correlation', ascending=False).head(top_n + 1)  # +1, чтобы исключить само аниме
    df_similar = df_similar[df_similar.index != anime_name]  # исключаем само аниме
    
    # применяем порог корреляции
    df_filtered = df_similar[df_similar['correlation'] >= threshold]
    
    # проверка есть ли похожие аниме
    if df_filtered.empty:
        print(f"Нет похожих аниме для '{anime_name}' с порогом корреляции выше {threshold}.")
        return pd.DataFrame()
    
    return df_filtered.head(top_n)
```

```python
similar_anime = same_anime("bleach", threshold=0.2)
print(similar_anime)
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
pivot_table = df_sample.pivot_table(
    index='user_id',
    columns='name',
    values='user rating',
).fillna(0)
```

```python
pivot_table
```

```python
# создание разряжённой матрицы на основе таблицы
sparse_matrix = csr_matrix(pivot_table.values)
```

```python
# размерность матрицы
sparse_matrix.shape
```

```python
# создание модели ближайших соседей
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(sparse_matrix);
```

```python
# функция для рекомендаций аниме
def get_recommendations(anime_name, n_recommendations=10):
    model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
    model_knn.fit(sparse_matrix);
    # проверка, существует ли аниме в таблице
    if anime_name not in pivot_table.columns:
        return f"Аниме {anime_name} не найдено."

    anime_index = pivot_table.columns.get_loc(anime_name)

    # вектор для аниме и преобразование его в массив с нужной формой
    anime_vector = sparse_matrix[anime_index, :].toarray().reshape(1, -1)

    # получаем ближайших соседей
    distances, indices = model_knn.kneighbors(anime_vector, n_neighbors=n_recommendations + 1)

    # получаем рекомендации
    recommended_anime = []
    for i in range(1, len(distances.flatten())):
        idx = indices.flatten()[i]
        if idx < len(pivot_table.columns):  # проверка, что индекс в пределах
            recommended_anime.append((pivot_table.columns[idx], distances.flatten()[i]))

    return recommended_anime
```

```python
anime_name = 'toradora'
recommendations = get_recommendations(anime_name)

# Проверяем, является ли recommendations списком
if isinstance(recommendations, list):
    print(f"Рекомендации для аниме {anime_name}:")
    for rec in recommendations:
        print(f"{rec[0]}")
else:
    # Выводим сообщение об ошибке, если recommendations не является списком
    print(recommendations)
```

### Косинусное сходство

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
# вычисление косинусного сходства
similarity_score = cosine_similarity(pivot_table.T)
```

```python
# создание перевёрнутой таблицы
pivot_table_T = pivot_table.T
```

```python
# функция для рекомендаций аниме по принципу косинусного сходства
def cos_recommend_anime(anime_name):
    try:
        # поиск индекса для указанного названия аниме в pivot_table_T
        index = np.where(pivot_table_T.index == anime_name)[0][0]
    except IndexError:
        print(f"'{anime_name}' не найдено")
        return []
    
    # получение списка схожих аниме
    similar_anime = sorted(
        list(enumerate(similarity_score[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:11]  # берем топ-10 схожих аниме, исключая первое (само себя)
    
    data = []
    
    for i in similar_anime:
        temp_df = df_sample[df_sample['name'] == pivot_table_T.index[i[0]]]
        if not temp_df.empty:
            data.extend(list(temp_df.drop_duplicates('name')['name'].values))
        else:
            print(f"Данные для аниме с индексом {i[0]} не найдены.")
    
    return data
```

```python
anime_name = "zombie loan"
recommended_animes = cos_recommend_anime(anime_name)
print(f"Рекомендованные аниме по типу {anime_name}:")
for i in recommended_animes:
    print("- " + i)
```

```python
anime_name = "naruto"
recommended_animes = cos_recommend_anime(anime_name)
print(f"Рекомендованные аниме по типу {anime_name}:")
for i in recommended_animes:
    print("- " + i)
```

###  Косинусное сходство по жанру

```python
# функция для рекомендаций аниме по жанру
def recommend_by_genre(genre_name, top_n=10):
    genre_df = df_sample[df_sample['genre'].str.contains(genre_name, case=False, na=False)]
    
    if genre_df.empty:
        print(f"Нет аниме с жанром '{genre_name}' в базе данных.")
        return []
    
    # удаление дубликатов названий и сортировка по рейтингу
    top_anime_by_genre = genre_df.drop_duplicates(subset='name').sort_values(by='rating', ascending=False).head(top_n)
    
    # создание списка с названиями
    recommendations = list(top_anime_by_genre['name'].values)
    
    return recommendations
```

```python
anime_genre = "action"
recommended_anime_genre = recommend_by_genre(anime_genre)
print(f"Лучшие аниме жанра {anime_genre}:")
for i in recommended_anime_genre:
    print("- " + i)
```
